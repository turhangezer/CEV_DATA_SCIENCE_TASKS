```python
def is_prime(num):
  if (num<2):
    return False
  elif (num==2):
    return True
  elif (num%2 == 0):
    return False
  i=3
  while (i<=num/2):
    if (num%i==0):
      return False
    i=i+1
  return True
```

```python
def first_100_prime():
    print('2')
    for i in range (3,542,2):
        if is_prime(i):
            print(i)
```
